## Tipos de Dados Booleanos

- Tem apenas dois valores válidos
    - <font color='green'>**True**</font>
    - <font color='green'>**False**</font>
- Você pode criar uma variável com os valores booleanos

```python
# criando duas variaveis com True e False
verdade = True
mentira = False

print(f'Valor da variável verdade: {verdade}')
print(f'Valor da variável mentira: {mentira}')
print('*************************************')
print(f'Tipo das duas variáveis: \nverdade: {type(verdade)} \nmentira: {type(mentira)}')
```

- Os valores booleanos devem ser escritos com letra inicial maiúscula e não precisam estar entre aspas

```python
# colocando com letra minúscula, o Python retorna um erro
verdade_minuscula = true
```

```python
# colocando entre aspas, o Python entende como uma string normal
verdade_string = 'True'
print(f'Valor da variável verdade_string: {verdade_string}')
print(f'Tipo da variável verdade_strin: {type(verdade_string)}')
```

```python
# tentando criar uma variavel com nome True
True = 2
```

> **ASPECTOS QUE VALEM A PENA SABER**
- Os tipos Booleanos são um subtipo do tipo inteiro
- Veja o que acontece quando transformamos um tipo booleano em inteiro
- Isso será útil quando quisermos contar quantos valores True ou False existem em uma certa estrutura de dado

```python
# usando a função int() em valores booleanos
verdade = True
mentira = False
print(f'Valor inteiro de verdade: {int(verdade)}')
print(f'Valor inteiro de mentira: {int(mentira)}')
```

> **ASPECTOS QUE VALEM A PENA SABER**
- Qualquer objeto do Python pode ser testado se é True ou False e depois usado numa condicional
- Temos uma função built-in do Python que transforma outros valores em um valor booleano
    - Essa função é a bool(): [link documentação - class bool([x])](https://docs.python.org/3/library/functions.html#bool)

```python
# constantes consideradas Falsas
const_none = None
const_false = False # esse valor nos ja vimos
print(f'Booleano de uma variável com valor None: {bool(const_none)}')

# qualquer valor numerico zero e considerado False
print(f'Booleano de um numérico zero: {bool(10 - 10)}')
print(f'Booleano de um numérico zero: {bool(0.0)}')
print(f'Booleano de um numérico zero: {bool(0)}')

# qualquer sequencia ou coleção vazia
lista_vazia = []
string_vazia = ''
print(f'Booleano de uma lista vazia: {bool(lista_vazia)}')
print(f'Booleano de uma string vazia: {bool(string_vazia)}')
```

## Operadores de Comparação e Expressões Booleanas

- Os OPERADORES DE COMPARAÇÃO fazem a comparação entre dois valores e retornam <font color='green'>True</font> ou <font color='green'>False</font>, dependendo do que foi fornecido para eles
- Qualquer expressão que pode ser <font color='green'>True</font> ou <font color='green'>False</font> é chamada de EXPRESSÃO BOOLEANA

```python
# variaveis
a = 100
b = 50
c = a / b
d = 2
e = 'ISTO É UM TEXTO'
f = 'isto é um texto'
g = b * 2

# operador <
print(f'Valor booleano da operação a < b: {bool(a < b)}')

# operador <=
print(f'Valor booleano da operação c <= d: {bool(c <= d)}')

# operador >
print(f'Valor booleano da operação a > b: {bool(a > b)}')

# operador >=
print(f'Valor booleano da operação a >= b: {bool(a >= b)}')

# operador ==
print(f'Valor booleano da operação a == b: {bool(a == b)}')
print(f'Valor booleano da operação a == g: {bool(a == g)}')
print(f'Valor booleano da operação e == f: {bool(e == f)}')

# operador !=
print(f'Valor booleano da operação e != f: {bool(e != f)}')
```

## Operadores Lógicos(Booleanos)
- São usados para comparar valores booleanos

> **OPERADOR <font color='green'>AND</font>**

TABELA VERDADE DO OPERADOR <font color='green'>AND</font>

```python
# variaveis
a = 100
b = 50
c = 10
 
print(f'Valor booleano da expressão (a < b) and (c > b): {(a < b) and (c > b)}')
print(f'Valor booleano da expressão (a > b) and (c < b): {(a > b) and (c < b)}')
```

> **OPERADOR <font color='green'>OR</font>**

TABELA VERDADE DO OPERADOR <font color='green'>OR</font>

```python
# variaveis
a = 100
b = 50
c = 10
 
print(f'Valor booleano da expressão (a < b) or (c < b): {(a < b) or (c < b)}')
print(f'Valor booleano da expressão (a < b) or (c > b): {(a < b) or (c > b)}')
```

> **OPERADOR <font color='green'>NOT</font>**

TABELA VERDADE DO OPERADOR <font color='green'>NOT</font>

```python
# variaveis
a = 100
b = 50
c = 10
 
print(f'Valor booleano da expressão (a < b) or (c < b): {not((a < b) or (c < b))}')
print(f'Valor booleano da expressão (a < b) or (c > b): {not((a < b) or (c > b))}')
```

## Condicionais
- Usamos os valores True ou False de expressões booleanas junto com a condicional IF
- O IF "decide" o que fazer dependendo do resultado da expressão booleana passada para ela
- A expressão booleana nesse contexto é chamada de CONDIÇÃO

- No exemplo abaixo, **SE** a condição/expressão booleana **ano_nascimento menor do que ou igual a 2000** for **True/Verdadeira**, o programa irá printar na tela a frase "Nasceu antes de 2000"

```python
"""
v1:
Exemplo de programa que pergunta o ano de Nascimento
e printa na tela caso ele tenha nascido antes de 2000
"""

ano_nascimento = int(input('Entre com o seu ano de nascimento: '))
if ano_nascimento < 2000:
    print('Nasceu antes de 2000')
```

- No programa acima, caso a **condição ano_nascimento menor do que ou igual a 2000** for **False**, nada ocorre
- Usamos a execução alternativa **ELSE** nesses casos

```python
"""
v2:
Exemplo de programa que pergunta o ano de Nascimento
e fala se ele nasceu antes de 2000 ou depois de 2000
"""

ano_nascimento = int(input('Entre com o seu ano de nascimento: '))
if ano_nascimento < 2_000:
    print('Nasceu antes de 2000')
else:
    print('Nasceu depois de 2000')
```

- Usamos **ELIF (CONDIÇÃO)** para colocar condições alternativas no meio da execução do CONTROLE DE FLUXO

```python
"""
v3:
Exemplo de programa que pergunta o ano de Nascimento
e fala se ele nasceu antes de 2000, em 2000 ou depois de 2000
"""

ano_nascimento = int(input('Entre com o seu ano de nascimento: '))
if ano_nascimento < 2_000:
    print('Nasceu antes de 2000')
elif ano_nascimento == 2_000:
    print('Nasceu exatamente em 2000')
else:
    print('Nasceu depois de 2000')
```

## Loop WHILE
- Repete um bloco de código

```python
"""
Verifica se a senha digitada esta correta.
Caso nao esteja, coloca um aviso na tela.
"""

password = 'rogerio2019'

password_digitado = input('Entre com a sua senha: ')

while password_digitado != password:
    print('Você digitou a senha errada. Por favor, digite a senha novamente.')
    password_digitado = input('\nEntre com a sua senha: ')

print('Parabéns. Você digitou a senha correta')
```

- Usando **BREAK** para escapar de um **LOOP WHILE**

```python
"""
Verifica se a senha digitada esta correta.
Caso nao esteja, coloca um aviso na tela.
Caso esteja correta, usa BREAK para sair do LOOP WHILE.
"""
password = 'rogerio2019'
while True:
    password_digitado = input('Digite sua senha: ')

    if password_digitado == password:
        print('Você acessou sua conta')
        break
    else:
        print('Você digitou a senha errada. Por favor, digite novamente.\n')
```

- Usando **CONTINUE** para voltar ao começo do **LOOP WHILE**

```python
"""
Verifica se voce esta digitando do Brasil.
Verifica se voce e o Rogerio.
Pede sua senha.
"""

while True:
    pais = input('De qual país você está digitando? ')
    nome = input('Qual o seu nome? ')
    password = input('Qual a sua senha? ')
    
    if pais != 'Brasil' or nome != 'Rogerio' or password != 'rogerio2019':
        print('Alguma das opções está errada. Digite todas novamente.')
        continue
    else:
        print('Access granted.')
        break
```

## Loop FOR
- Usamos quando queremos repetir um bloco de código um determinado número de vezes
- Podemos usar a função range() para especificar o número de vezes que queremos

```python
# printando numeros de 0 a 3
for i in range(3):
    print(i)
```

```python
# printando numeros de 1 a 5
for i in range(1, 6):
    print(i)
```

```python
# printando numeros de 1 a 10, pulando de 2 em 2
for i in range(1, 11, 2):
    print(i)
```

```python
# printando numeros de 4 a 1
for i in range(4, 0, -1):
    print(i)
```
